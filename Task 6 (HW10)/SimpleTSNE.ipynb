```python
from sklearn import datasets
import json
from openTSNE import TSNE
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
x, y = iris["data"], iris["target"]

embedding = TSNE().fit(x)
```

```python
data = pd.DataFrame(embedding, columns=["x","y"])
data["label"] = y
```

```python
data.to_csv('data.csv', index = False)
```

```python
# TODO add the Target Information into  your json file
with open('data.json', 'w') as f:
    json.dump(np.array(embedding).tolist(), f)
```

```python
sns.set_context("notebook", font_scale=1.1)
sns.set_style("ticks")

sns.lmplot(x='Comp1',
           y='Comp2',
           data=data,
           fit_reg=False,
           legend=True,
           size=9,
           hue='Label',
           scatter_kws={"s":200, "alpha":0.3}
          )

plt.title("TSNE", weight='bold').set_fontsize('14')
plt.xlabel('Comp 1')
plt.ylabel('Comp 2')
```

```python
from sklearn import datasets
import json
# from openTSNE import TSNE
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from umap import UMAP

iris = datasets.load_iris()
x, y = iris["data"], iris["target"]

embedding = UMAP().fit_transform(x)

data = pd.DataFrame(embedding, columns=["x","y"])
data["label"] = y

data.to_csv('data_umap.csv', index = False)

# TODO add the Target Information into  your json file
# with open('data.json', 'w') as f:
#     json.dump(np.array(embedding).tolist(), f)

sns.set_context("notebook", font_scale=1.1)
sns.set_style("ticks")

sns.lmplot(x='x',
           y='y',
           data=data,
           fit_reg=False,
           legend=True,
           size=9,
           hue='label',
           scatter_kws={"s":200, "alpha":0.3}
          )

plt.title("UMAP", weight='bold').set_fontsize('14')
plt.xlabel('Comp 1')
plt.ylabel('Comp 2')
```
